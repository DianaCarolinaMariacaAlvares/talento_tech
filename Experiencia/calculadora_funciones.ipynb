### ***Calculadora con funciones***

```python
import math

# Definición de funciones

# Función de presentación de menú
def presentar_menu():
    # Definición del menú
    menu = ["1. Sumar",
            "2. Restar",
            "3. Multiplicar",
            "4. Dividir",
            "5. Potenciar",
            "6. Radicar",
            "7. Logaritmo",
            "8. Seno",
            "9. Coseno",
            "10. Tangente",
            "11. Salir"]
    # Impresión del menú
    print()
    print("MENU\n")
    for i in menu:
        print(i)
    print()
    # Retorno de cantidad de opciones
    return len(menu)

# Función de lectura de opción
def ingresar_opcion(cantidad_opciones):
    # Lectura de opción
    opcion_valida = False
    while opcion_valida == False:
        try:
            opcion = int(input("Ingrese el número de la opción que desee ejecutar: "))
        except:
            print("Opción inválida, intente de nuevo")
        else:
            if opcion > cantidad_opciones:
                print("Opción inválida, intente de nuevo")
            else:
                opcion_valida = True
    return opcion

# Función de lectura de numeros
def ingresar_numeros(opcion):
    # Establecimiento de terminos para los números a solicitar
    match opcion:
        case 4:
            nombre_numero_uno = "el dividendo"
            nombre_numero_dos = "el divisor"
        case 5:
            nombre_numero_uno = "la base"
            nombre_numero_dos = "el exponente"
        case 6:
            nombre_numero_uno = "el radicando"
            nombre_numero_dos = "el indice"
        case 7:
            nombre_numero_uno = "el argumento"
            nombre_numero_dos = "la base"
        case _:
            nombre_numero_uno = "el primer número"
            nombre_numero_dos = "el segundo número"
            nombre_numero = "el ángulo"
    
    # Determinación de texto a presentar y lectura de datos para las operaciones
    if opcion < 8:
        mensaje_uno = "Ingrese " + nombre_numero_uno + ": "
        mensaje_dos = "Ingrese " + nombre_numero_dos + ": "
        numero_uno = None
        while numero_uno == None:
            try:
                numero_uno = float(input(mensaje_uno))
            except:
                print("Número inválido, intente de nuevo")

        numero_dos = None
        while numero_dos == None:
            try:
                numero_dos = float(input(mensaje_dos))
            except:
                print("Número inválido, intente de nuevo")
        return numero_uno, numero_dos
    else: 
        mensaje = "Ingrese " + nombre_numero + " en grados: "
        numero_uno = None
        while numero_uno == None:
            try:
                numero_uno = float(input(mensaje))
            except:
                print("Número inválido, intente de nuevo")
        return numero_uno, None
    
# Establecimiento de funciones para cada operación
def calcular_suma(numero_uno,numero_dos):
    return numero_uno + numero_dos
def calcular_resta(numero_uno,numero_dos):
    return numero_uno - numero_dos
def calcular_multiplicacion(numero_uno,numero_dos):
    return numero_uno * numero_dos
def calcular_division(numero_uno,numero_dos):
    if numero_dos == 0:
        print(f"El denominador es 0, no se puede realizar la división entre {numero_uno} y {numero_dos}")
        return "inválido"
    else:
        return numero_uno / numero_dos
def calcular_potencia(numero_uno,numero_dos):
    return math.pow(numero_uno,numero_dos)
def calcular_raiz(numero_uno,numero_dos):
    return math.pow(numero_uno,(1/numero_dos))
def calcular_logaritmo(numero_uno,numero_dos):
    if numero_uno > 0 and (numero_dos > 0 and numero_dos != 1):
        return math.log(numero_uno,numero_dos)
    else:
        return "inválido"
def calcular_seno(numero):
    return math.sin(math.radians(numero))
def calcular_coseno(numero):
    return math.cos(math.radians(numero))
def calcular_tangente(numero):
    if round(math.cos(math.radians(numero)),10) == 0:
        print(f"No existe tangente del valor {numero}")
        return "inválida"
    else:
        return math.tan(math.radians(numero))

# Definición de opciones
def opcion_uno(numero_uno, numero_dos):
    print(f"Vamos a sumar los números {numero_uno} y {numero_dos}")
    suma = calcular_suma(numero_uno, numero_dos)
    print(f"El resultado de la suma es: {suma}")

def opcion_dos(numero_uno, numero_dos):
    print(f"Vamos a restar los números {numero_uno} y {numero_dos}")
    resta = calcular_resta(numero_uno, numero_dos)
    print(f"El resultado de la resta es: {resta}")

def opcion_tres(numero_uno, numero_dos):
    print(f"Vamos a multiplicar los números {numero_uno} y {numero_dos}")
    producto = calcular_multiplicacion(numero_uno, numero_dos)
    print(f"El producto de la multiplicación es: {producto}")

def opcion_cuatro(numero_uno, numero_dos):
    print(f"Vamos a dividir {numero_uno} entre {numero_dos}")
    cociente = calcular_division(numero_uno, numero_dos)
    print(f"El cociente de la división es: {cociente}")

def opcion_cinco(numero_uno, numero_dos):
    print(f"Vamos a calcular la potencia con base {numero_uno} y exponente {numero_dos}")
    potencia = calcular_potencia(numero_uno, numero_dos)
    print(f"La potencia es: {potencia}")

def opcion_seis(numero_uno, numero_dos):
    print(f"Vamos a calcular la raíz con radicado {numero_uno} e índice {numero_dos}")
    raiz = calcular_raiz(numero_uno, numero_dos)
    print(f"La raíz es: {raiz}")

def opcion_siete(numero_uno, numero_dos):
    print(f"Vamos a calcular el logaritmo del argumento {numero_uno} en base {numero_dos}")
    logaritmo = calcular_logaritmo(numero_uno, numero_dos)
    print(f"El logaritmo es: {logaritmo}")

def opcion_ocho(numero):
    print(f"Vamos a calcular el seno de {numero}")
    seno = calcular_seno(numero)
    print(f"El seno es: {round(seno,10)}")

def opcion_nueve(numero):
    print(f"Vamos a calcular el coseno de {numero}")
    coseno = calcular_coseno(numero)
    print(f"El coseno es: {round(coseno,10)}")

def opcion_diez(numero):
    print(f"Vamos a calcular la tangente de {numero}")
    tangente = calcular_tangente(numero)
    if tangente == "inválida":
        print(f"La tangente es: {tangente}")
    else:
        print(f"La tangente es: {round(tangente,10)}")

def default():
    print("Opción no válida")

# Procesamiento de opciones y operaciones
def procesar_y_mostrar_operaciones(opcion,numero_uno,numero_dos):
    match opcion:
        case 1:
            opcion_uno(numero_uno,numero_dos)
        case 2:
            opcion_dos(numero_uno,numero_dos)
        case 3:
            opcion_tres(numero_uno,numero_dos)
        case 4:
            opcion_cuatro(numero_uno,numero_dos)
        case 5:
            opcion_cinco(numero_uno,numero_dos)
        case 6:
            opcion_seis(numero_uno,numero_dos)
        case 7:
            opcion_siete(numero_uno,numero_dos)
        case 8:
            opcion_ocho(numero_uno)
        case 9:
            opcion_nueve(numero_uno)
        case 10:
            opcion_diez(numero_uno)
        case _:
            default()
        
# Función de trabajo general
def llamar_funciones():
    opcion = None
    cantidad_opciones = presentar_menu()
    while opcion != 11:
        opcion = ingresar_opcion(cantidad_opciones)
        if opcion == 11:
            print("SALIENDO...")
        else:
            numero_uno, numero_dos = ingresar_numeros(opcion)
            procesar_y_mostrar_operaciones(opcion,numero_uno,numero_dos)

# Ejecución del programa
llamar_funciones()
```
