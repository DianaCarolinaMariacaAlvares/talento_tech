```python
# Ejemplo lista
lista_animales = ["Perro", "Gato", "Ratón", "Caballo", "Serpiente", "Tigre", "Tiburón"] 
# Ejemplo tupla
tupla_aves = ("Gorrión", "Loro", "Tórtola", "Gavilán", "Azulejo", "Mirla", "Urraca")
# Ejemplo conjunto
conjunto_mascotas = {"Perro", "Gato", "Conejo", "Tortuga", "Loro", "Pez"}
# Ejemplo directorio
directorio_mascota1 = {"Nombre":"Luis", "Especie":"Gato", "Edad":"3 años"}
directorio_mascota2 = {"Nombre":"Thaner", "Especie":"Perro", "Edad":"7 años"}
directorio_mascota3 = {"Nombre":"Manchas", "Especie":"Perro", "Edad":"2 meses"}
directorio_mascotas = {"Mascota_1":directorio_mascota1,"Mascota_2":directorio_mascota2,"Mascota_3":directorio_mascota3}

# impresión lista
print("Lista de animales\n")
for i in lista_animales:
    print (i)
print()

# impresión tupla
print("Tupla aves\n")
for i in tupla_aves:
    print (i)
print()

# impresión conjunto
print("Conjunto mascotas\n")
for i in conjunto_mascotas:
    print (i)
print()

# comparación entre lista y conjunto
print("Animales que se pueden considerar mascotas\n")
for i in conjunto_mascotas:
    for j in lista_animales:
        if i == j:
            print(i)
print()

# impresión llaves de directorio
print("Directorio datos de mascota: llaves\n")
for i in directorio_mascota1.keys():
    print (i)
print()

# impresión directorio
print("Directorio de mascotas\n")
for i, j in directorio_mascotas.items():
    print (i,j)
print()

# impresión directorio ciclos anidados
print("Directorio de mascotas\n")
for i, j in directorio_mascotas.items():
    print(i)
    for k,l in j.items():
        print(f"{k}: {l}")    
    print()
```
